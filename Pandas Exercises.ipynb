** How many banks has the State Bank of Texas acquired? How many of them were actually in Texas?**

```python
# CODE HERE
Bank_of_Texas = banks[banks['Acquiring Institution']=='State Bank of Texas']
print(' the State Bank of Texas acquired', len(Bank_of_Texas), 'Banks in total')
print(' the State Bank of Texas acquired', len(Bank_of_Texas[Bank_of_Texas['ST'] == 'TX']), 'Banks in Texas')
banks[banks['Acquiring Institution']=='State Bank of Texas'].head()
```

** What is the most common city in California for a bank to fail in?**

```python
# CODE HERE
cali = banks[banks['ST']=='CA']
cali.head()
cali_city = cali.groupby('City')

print('The most common city for a bank to fail in is', cali_city['City'].count().sort_values(ascending=False).head(1))
```

** How many failed banks don't have the word "Bank" in their name? **

```python
# CODE HERE
len(banks[~banks['Bank Name'].str.contains("Bank")])
```

** How many bank names start with the letter 's' ? **

```python
# CODE HERE
len(banks[banks['Bank Name'].str.startswith('S')])
```

** How many CERT values are above 20000 ? **

```python
# CODE HERE
len(banks[banks['CERT'] > 20000])
```

** How many bank names consist of just two words? (e.g. "First Bank" , "Bank Georgia" )**

```python
# CODE HERE
len(banks[banks['Bank Name'].str.count(' ') + 1 == 2])
```

**Bonus: How many banks closed in the year 2008? (this is hard because we technically haven't learned about time series with pandas yet! Feel free to skip this one!**
